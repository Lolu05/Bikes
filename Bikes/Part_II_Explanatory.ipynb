## Investigation Overview

> In this investigation, I wanted to look at the influence of the features on one another. The main focus was on the user types, weekdays, distance covered, age, gender and duration minutes.

## Dataset Overview

> The actual data consisted of attributes of over 18,000 trips. The attributes included user type, gender, start time, end time, duration minutes and so on. Some data points were removed from the analysis due to inconsistencies or missing information and outliers. The cleaned dataset was used for the explanatory data analysis.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
bikes = pd.read_csv('bikes_new.csv')
```

## Distribution of the duration minutes

> The distribution is positively skewed. Most trips took about 0 to 20 minutes.



```python
bin_edges = np.arange(0, 100, 5)

plt.hist(data = bikes, x = 'duration_min', bins = bin_edges, color="green")
plt.title("Trip Duration Histogram", y=1.03, fontsize=15, fontweight='semibold')
plt.xlabel('Duration in minutes')
plt.ylabel('Number of rides');
```

## Distribution of distance traveled

> This is to know the distance covered by most users. The distribution is positively skewed and majority covered between the range of 0.5km to 2km.

```python
bin_edges = np.arange(0, 6, 0.3)
plt.hist(data = bikes, x = 'distance_km', bins = bin_edges,  color="green")
plt.title("Distance Traveled Histogram", y = 1.06, fontsize = 15, fontweight='semibold')
plt.xlabel('Distance in KM')
plt.ylabel('Number of Bike Trips');
```

## User types and weekdays

> This is to know the weekday usage of the users. Both customers and subscribers take trips mostly on Thursdays. However, it seems most of the subscribers are working class because compared to customers, they do not really take trips during weekends.

```python
order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
bi1= sns.catplot(data=bikes, x='start_time_weekday', col="user_type", kind='count', sharey = False, order = order, color = 'green')
bi1.set_axis_labels("Weekday", "Number of Bike Trips")
bi1.set_titles("{col_name}")
bi1.fig.suptitle('Weekday Usage of the Bikeshare System by User Type', y=1.03, fontsize=15, fontweight='semibold')
```

## User types and Distance

> Customers covered more distance than subscribers. The average distance covered by customers was 1.75km while that of subscribers was 1.50km

```python
bi2= sns.catplot(data=bikes, x='distance_km', y="user_type", kind='bar', sharey = False, color = 'green', orient = 'h')
bi2.set_axis_labels("Distance covered", "User type")
bi2.set_titles("{col_name}")
bi2.fig.suptitle('Distance covered by User Type', y=1.03, fontsize=15, fontweight='semibold')
```

## User types and Duration minutes

> Average trip was taken by customers, although subscribers seem to be more than customers in terms of count.

```python
# looking at the type of users with average trip duration
plt.figure(figsize = (15,7))
bi3 = sns.catplot(x = "user_type", y = "duration_min", data = bikes, kind = 'bar', color = 'green')
bi3.set_axis_labels("Trip duration", "User type")
bi3.set_titles("{col_name}")
bi3.fig.suptitle('Trip Duration by User Type', y=1.03, fontsize=15, fontweight='semibold');
```

## Weekday Usage of the Bikeshare system 

> The male gender had the highest usage. However, the proportion of female customers that ride on weekends are more than the proportion of female subscribers.

```python
order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
mu2= sns.catplot(data=bikes, x='start_time_weekday', col="user_type", hue='member_gender', kind='count', sharey = False,
                order = order, color = 'green')
mu2.set_axis_labels("Weekday", "Number of Bike Trips")
mu2._legend.set_title('Gender')
mu2.set_titles("{col_name}")
mu2.fig.suptitle('Weekday Usage of the Bikeshare System by User Type and Gender', y=1.03, fontsize=15, fontweight='semibold');
```

## Distance covered and gender by users

> Customers generally covered more distance. The male subscribers did not really cover as much distance as the females. Other users had the highest distance covered.

```python
colors = ['#3CB371', '#808080']
plt.title('Distance covered by each user type and gender', fontsize=15, fontweight='semibold')
sns.pointplot(data=bikes, x="member_gender", y="distance_km", hue="user_type", palette = colors, dodge=True);
```

## Age of Users by gender

> Male customers were mostly 32 years, female customers were 31 years while others are 35 years. Looking at the subscribers, the same thing is applicable to them, except that the age of other subscribers was higher than that of other customers considering the boxplot


```python
fig,ax1 = plt.subplots(figsize = (8,5))
colors = ['#3CB371', '#FFA701' ,'#808080']
plt.title("Disparity of the age of the users' gender ", fontsize=15, fontweight='semibold')
sns.boxplot (ax = ax1, x = 'user_type', y = 'member_age', hue = 'member_gender', data = bikes, palette = colors );
```
